```python
Number=int(input("Enter the Number->"))
First=0                                         
Second=1                                         
if Number<=0:
    print("Enter greater than 0")
else:
    print(First,Second,end=' ')
    for i in range(2,Number):
        Next=First+Second
        First=Second
        Second=Next
        if Next>50:
            break
        print(Next,end=" ")
```
